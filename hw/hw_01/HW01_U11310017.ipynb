```python


## 匯入 obspy
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
from obspy import Stream, Trace

## 從 IRIS 下載地震波形資料
client = Client("IRIS")
t = UTCDateTime("2020-01-01T00:00:00")
st = client.get_waveforms(network="IU", station="ANMO", location="00", channel="BHZ",
                          starttime=t, endtime=t + 60*5)  # 5分鐘資料
st.plot()
```

```python
# 使用 pygmt 畫出台灣地圖
import pygmt

fig = pygmt.Figure()
fig.basemap(region=[119, 123, 21, 26], projection="M6i", frame=True)
fig.coast(shorelines=True, land="lightgray", water="skyblue")
fig.show()
```
